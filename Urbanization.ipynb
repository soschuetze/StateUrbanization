```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv('urbanization-census-tract.csv', header=0)
df.head()
```

```python
dfIndex = df[['state','urbanindex']]
dfIndex.head()
```

```python
dfAvg = dfIndex.groupby(['state'],as_index=False ).mean()
dfAvg
```

```python
dfDec = dfAvg.sort_values(by='urbanindex', ascending=False)
dfDec
```

```python
import seaborn as sns
sns.set(style="ticks", color_codes=True)
sns.catplot(x="state", y="urbanindex", kind="bar", data=dfDec);
```
